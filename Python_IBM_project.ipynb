```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Splitting dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Dictionary of models
models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}
results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.bar(results.keys(), results.values(), color='lightblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}
# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# Select the best performing model
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\nBest model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print("Saved best model as best_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np

# --- Load Model and Define Column Structure ---

try:
    model = joblib.load("best_model.pkl")
except FileNotFoundError:
    st.error("Error: The model file 'best_model.pkl' was not found.")
    st.stop()

# ============================ IMPORTANT: FIX THIS SECTION ============================
# You MUST update these lists to match the 13 features your model was trained on.
# The following is just an EXAMPLE that adds up to 13 features.

# EXAMPLE: Let's assume your model used these 3 numerical features.
NUMERICAL_FEATURES = ['age', 'educational-num', 'hours-per-week']

# EXAMPLE: ...and was trained on these 6 'workclass' categories.
WORKCLASS_COLS = ['workclass_Federal-gov', 'workclass_Local-gov', 'workclass_Private', 'workclass_Self-emp-inc', 'workclass_Self-emp-not-inc', 'workclass_State-gov']

# EXAMPLE: ...and these 4 'marital-status' categories. (5 in the UI, but maybe only 4 in training data)
MARITAL_COLS = ['marital-status_Divorced', 'marital-status_Married-civ-spouse', 'marital-status_Never-married', 'marital-status_Separated']
# Note: 'Widowed' is left out in this example to make the total 13. Your model might be different.

# EXAMPLE: The 'occupation' and 'experience' features are NOT used in this example.
OCCUPATION_COLS = [] # Left empty if the model didn't use this feature
# Total features = 3 (numerical) + 6 (workclass) + 4 (marital) = 13.

# This final list MUST contain your model's exact 13 column names in the correct order.
MODEL_COLUMNS = NUMERICAL_FEATURES + WORKCLASS_COLS + MARITAL_COLS + OCCUPATION_COLS
# ====================================================================================


# --- Preprocessing Function ---

def preprocess_input(df):
    processed_df = df.copy()

    # 1. Map Education Level
    education_map = {
        "Bachelors": 13, "Masters": 14, "PhD": 16,
        "HS-grad": 9, "Assoc": 11, "Some-college": 10
    }
    processed_df['educational-num'] = processed_df['education-str'].map(education_map).fillna(0)
    processed_df = processed_df.drop('education-str', axis=1)

    # 2. One-Hot Encode
    categorical_features = ['workclass', 'marital-status', 'occupation']
    # We create dummies for all, but will only keep the ones defined in MODEL_COLUMNS
    processed_df = pd.get_dummies(processed_df, columns=categorical_features)

    # 3. Align Columns with the Model's Training Columns
    # This ensures the final DataFrame has exactly the 13 columns the model expects.
    processed_df = processed_df.reindex(columns=MODEL_COLUMNS, fill_value=0)

    return processed_df


# --- Streamlit App UI (No changes needed here) ---

st.set_page_config(page_title="Employee Salary Prediction", page_icon="ðŸ’µ", layout="centered")
st.title("ðŸ’µ Employee Salary Prediction: Using Machine Learning Algorithms")
st.markdown("Predict an employee's salary range (>50K or â‰¤50K).")

# Sidebar for User Inputs
st.sidebar.header("ðŸ§‘â€ðŸ’¼ Employee Profile")
age = st.sidebar.slider("Age", 18, 70, 35)
experience = st.sidebar.slider("Years of Experience", 0, 40, 10) # Note: This input might not be used by your model
hours_per_week = st.sidebar.slider("Hours per Week", 1, 99, 40)
education_str = st.sidebar.selectbox("Education Level", ["HS-grad", "Some-college", "Bachelors", "Masters", "PhD", "Assoc"])
workclass = st.sidebar.selectbox("Work Class", ["Private", "Self-emp-not-inc", "Self-emp-inc", "Federal-gov", "Local-gov", "State-gov"])
marital_status = st.sidebar.selectbox("Marital Status", ["Married-civ-spouse", "Never-married", "Divorced", "Separated", "Widowed"])
occupation = st.sidebar.selectbox("Job Role", ["Prof-specialty", "Exec-managerial", "Sales", "Tech-support", "Craft-repair", "Other-service"]) # Note: This input might not be used

# Create the initial DataFrame
input_data = {
    'age': [age], 'experience': [experience], 'hours-per-week': [hours_per_week],
    'education-str': [education_str], 'workclass': [workclass],
    'marital-status': [marital_status], 'occupation': [occupation]
}
input_df = pd.DataFrame(input_data)

# Process the data for prediction
processed_display_df = preprocess_input(input_df)

st.write("### Processed Input for Model (13 Features)")
st.write(processed_display_df)

if st.sidebar.button("Predict Salary Class", use_container_width=True, type="primary"):
    # Use the correctly processed data for prediction
    prediction = model.predict(processed_display_df)
    prediction_proba = model.predict_proba(processed_display_df)

    # Display result
    st.subheader("ðŸ“Š Prediction Result")
    result_text = prediction[0]
    confidence = np.max(prediction_proba) * 100

    if ">" in str(result_text):
        st.success(f"**Predicted Salary Range:** {result_text} (Confidence: {confidence:.2f}%)")
    else:
        st.info(f"**Predicted Salary Range:** {result_text} (Confidence: {confidence:.2f}%)")




# --- Batch Prediction Section (with Debugging) ---
st.markdown("---")
st.markdown("### ðŸ—‚ï¸ Batch Prediction from CSV")
uploaded_file = st.file_uploader("Upload a CSV file with employee profiles", type="csv")

if uploaded_file is not None:
    try:
        # Step 1: Read the CSV
        batch_data = pd.read_csv(uploaded_file)
        st.write("Uploaded data preview:")
        st.dataframe(batch_data.head())

        # --- DEBUGGING ---
        st.info("1. Preprocessing the uploaded data...")
        processed_batch_df = preprocess_input(batch_data.rename(columns={'educational-num': 'education-str'}))

        st.info("2. Making predictions...")
        batch_preds = model.predict(processed_batch_df)

        # --- DEBUGGING: Check the raw predictions from the model ---
        st.info("3. Here are the first 5 raw predictions from the model:")
        st.write(batch_preds[:5])

        st.info("4. Adding the 'Predicted_Salary_Class' column to the DataFrame...")
        batch_data['Predicted_Salary_Class'] = batch_preds

        # --- DEBUGGING: Check the DataFrame's columns ---
        st.info("5. Columns in the DataFrame AFTER adding predictions:")
        st.write(batch_data.columns.tolist())

        # --- FINAL STEP: Display the result ---
        st.write("Predictions:")
        st.dataframe(batch_data.head())

        # Download button
        csv = batch_data.to_csv(index=False).encode('utf-8')
        st.download_button(
            "Download Predictions CSV",
            csv,
            file_name='predicted_salary_classes.csv',
            mime='text/csv'
        )

    except Exception as e:
        st.error(f"An error occurred during batch processing: {e}")
        st.exception(e) # This will print the full traceback of the error
```
